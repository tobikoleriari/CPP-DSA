4. Write a program using functions that determines whether a character input by a
user is uppercase or lower case.

```c++17
void isUpperCase(char c){
    if(isupper(c)){
        cout << c << " is uppercase"<<endl;
    }
    else {
        cout << c << " is not uppercase" << endl;
    }
}
```
